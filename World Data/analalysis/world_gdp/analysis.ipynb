```python
import pandas as pd

df = pd.read_csv('world_gdp_cleaned.csv')
df.head()
```

```python
df.drop(['Unnamed: 0','Age','%individuals using internet','Life expectancy at birth','Gini index','GDP growth (annual %)'],axis=1,inplace=True)
```

```python
scaled_list = df.select_dtypes('float').columns.to_list()
scaled_list
```

```python
df['Military expenditure (% of GDP)'] = df['Military expenditure (% of GDP)']*df['Current GDP ($)']/100
df
```

```python
df.rename({'Military expenditure (% of GDP)':'Military expenditure'},axis=1,inplace=True)
```

```python
#scale

df[scaled_list].apply(lambda x:x/x.max())
```
