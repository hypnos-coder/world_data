```python
import pandas as pd
df = pd.read_csv('world_education_cleaned.csv')
df
```

```python
df.drop(['Unnamed: 0','Age','GDP (current $)'],axis=1,inplace=True)
df
```

```python
zero_stat = (df==0).sum(axis=0).reset_index()
zero_stat = zero_stat[zero_stat[0]/len(df)>.55]
zero_stat['index'].to_list()
```

```python
df.drop(columns = zero_stat['index'].to_list(),axis = 1,inplace = True)
df
```

```python
columns_to_scale = df.select_dtypes('float').columns.to_list()

df[columns_to_scale] = df[columns_to_scale].apply(lambda x: x/x.max())
columns_to_scale
```

```python
df
```

```python
final_scale_list = [
    'Compulsory education, duration (years)',
    'Government expenditure on education, total (% of GDP)',
    'Government expenditure on education, total (% of government expenditure)',
    'Current education expenditure, secondary (% of total expenditure in secondary public institutions)',
]

```

```python
df['Weighted World Education'] = sum([df[final_scale_list[i]]*(i+1) for i in range(len(final_scale_list)-1,-1,-1)])
df
```

```python
import plotly.express as px
target = 'Weighted World Education'

# Create an animated choropleth map
fig = px.choropleth(
    data_frame=df,
    locations='Country Name',
    locationmode='country names',
    color=target,
    animation_frame='Year',
    color_continuous_scale='blues',
    range_color=(df[target].max(),df[target].min()),
    title='Time Series on World Map',
    labels={'Value': target}
)

# Customize the layout
fig.update_geos(showframe=False, showcoastlines=False, projection_type='orthographic',showocean=True, oceancolor="LightBlue",)
fig.update_layout(
    width=1000,  # Set the width of the figure
    height=800 , # Set the height of the figure
    transition_duration = 2000
)


# Display the plot
fig.show()
```
