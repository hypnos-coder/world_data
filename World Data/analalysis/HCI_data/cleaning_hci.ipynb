```python
def get_continent(row):
    try:
        if pd.notnull(row['Country Code']):
            # Use the country code if available
            country_info = pycountry.countries.get(alpha_3=row['Country Code'])
        else:
            # Use the country name as fallback
            country_info = pycountry.countries.get(name=row['Country Name'])

        if country_info is not None:
            # Retrieve the alpha-2 code
            alpha_2 = country_info.alpha_2

            # Retrieve the continent based on the alpha-2 code
            continent_info = pycountry.convert_country_alpha2_to_continent_code(alpha_2)
            
            if continent_info is not None:
                return continent_info
        return 'Unknown'
    except (KeyError, AttributeError):
        return 'Unknown'
```

```python
import pycountry_convert as pc

def country_to_continent(country_code):
    try:
        country_info = pycountry.countries.get(alpha_3=country_code)
        
        if country_info is not None:
                # Retrieve the alpha-2 code
            alpha_2 = country_info.alpha_2
            country_continent_code = pc.country_alpha2_to_continent_code(alpha_2)
            country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
            return country_continent_name
    except (KeyError, AttributeError):
        return 'Unknown'

# Example
country_C = 'SXM'
print(country_to_continent(country_C))
```

```python
# Assuming your dataset is a pandas DataFrame and the country column is named 'Country'
df['Continent']=df['Country Code'].apply(country_to_continent)
df
```
