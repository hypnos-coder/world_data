```python
import pandas as pd
import plotly.express as px


df = pd.read_csv('cleaned_governace_data.csv')#read the data to a data frame
df.head()
```

```python
df.drop(['Unnamed: 0', 'Age'],axis=1,inplace=True)# drop unnecessary columns (The column Unnamed seems to have been created during the conversion to csv)
```

```python
df.columns.to_list()
```

## Weight
for the upcoming analysis, I decided to put all the variable under the same umbrella by assigning weight to them and adding them up. This will give us 
one column that describe the dataset as a whole.
For the weight, I gave Chatgpt the list of column and receive a ranking, the rank are used for the weight but inversed.

```python
sorted_columns = [
 'Rule of Law Score',
 'Control of Corruption Score',
 'Government Effectiveness Score',
 'Political Stability Score',
 'Regulatory Quality Score',
 'Free Speech Score'
]# ranked from first to last list of column
```

```python
df['Weighted Governance'] = sum([df[sorted_columns[i]]*(i+1) for i in range(len(sorted_columns)-1,-1,-1)])# creation of the weighted column (notice the i+1, where i is the rank, the +1 is there becaus
# the Chatgpt did the ranking starting from zero)
```

```python
df
```

```python
target = 'Weighted Governance'


# Create an animated choropleth map
fig = px.choropleth(
    data_frame=df,
    locations='Country Name',
    locationmode='country names',
    color=target,
    animation_frame='Year',
    color_continuous_scale='YlOrRd_r',
    range_color=(df[target].max(),df[target].min()),
    title='Time Series on World Map',
    labels={'Value': target}
)

# Customize the layout
fig.update_geos(showframe=False, showcoastlines=False, projection_type='orthographic',showocean=True, oceancolor="LightBlue",)
fig.update_layout(
    width=1000,  # Set the width of the figure
    height=800 , # Set the height of the figure
    transition_duration = 2000
)


# Display the plot
fig.show()
```
