```python
import pandas as pd

df =  pd.read_csv('world_death_cleaned.csv')
df
```

```python
df.drop(['Unnamed: 0','Age'],axis=1,inplace=True)
df
```

```python
zero_stat = (df==0).sum(axis=0).reset_index()
zero_stat = zero_stat[zero_stat[0]/len(df)>.5]
zero_stat['index'].to_list()
```

```python
df.drop(columns = zero_stat['index'].to_list()+['GDP (current US$)'],axis = 1,inplace = True)
df
```

```python
columns_to_scale = df.select_dtypes('float').columns.to_list()

df[columns_to_scale] = df[columns_to_scale].apply(lambda x: x/x.max())
columns_to_scale
```

```python
final_scale_list = [
    'Life Expectancy at birth (total years)',
    'Mortality rate, infant (per 1,000 live births)',
    'Mortality rate, neonatal (per 1,000 live births)',
    'Mortality rate, under-5 (per 1,000 live births)',
    'Mortality from CVD, cancer, diabetes or CRD between ages 30 and 70 (%)',
    'Mortality rate attributed to household and ambient air pollution, age-standardized (per 100,000 population)',
    'Mortality rate attributed to unsafe water, unsafe sanitation and lack of hygiene (per 100,000 population)',
    'Mortality rate attributed to unintentional poisoning (per 100,000 population)',
    'Suicide mortality rate (per 100,000 population)',
    'Mortality cauased by road traffic injury (per 100,000 population)',
]

df['Weighted World Death'] = sum([-df[final_scale_list[i]]*(i+1) if 'Mortality' in final_scale_list[i] else df[final_scale_list[i]]*(i+1) for i in range(len(final_scale_list)-1,-1,-1)])
df
```

## plotly is then used to plot the data on a world map


```python
import plotly.express as px
target = 'Weighted World Death'

# Create an animated choropleth map
fig = px.choropleth(
    data_frame=df,
    locations='Country Name',
    locationmode='country names',
    color=target,
    animation_frame='Year',
    color_continuous_scale='YlOrRd_r',
    range_color=(df[target].max(),df[target].min()),
    title='Time Series on World Map',
    labels={'Value': target}
)

# Customize the layout
fig.update_geos(showframe=False, showcoastlines=False, projection_type='orthographic',showocean=True, oceancolor="LightBlue",)
fig.update_layout(
    width=1000,  # Set the width of the figure
    height=800 , # Set the height of the figure
    transition_duration = 2000
)


# Display the plot
fig.show()
```
